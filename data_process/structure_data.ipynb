```python
print("\nüîç Ki·ªÉm tra c√°c b·∫•t th∆∞·ªùng trong d·ªØ li·ªáu m·∫´u:\n")

anomalies = []

# 1. Ki·ªÉm tra InvoiceNo c√≥ d·∫°ng l·∫° kh√¥ng
print("1Ô∏è‚É£ InvoiceNo:")
no_inv = 0
for inv in df['InvoiceNo']:
    if pd.isna(inv):
        anomalies.append("‚ùå InvoiceNo c√≥ gi√° tr·ªã NaN")

        
if not anomalies:
    print("   ‚úì InvoiceNo c√≥ d·∫°ng b√¨nh th∆∞·ªùng (s·ªë ho·∫∑c s·ªë+C)")
else:
    for a in anomalies:
        print(f"   {a}")

# 2. Ki·ªÉm tra Quantity c√≥ √¢m kh√¥ng
print("\n2Ô∏è‚É£ Quantity:")
negative_qty = (df['Quantity'] < 0).sum()
if negative_qty > 0:
    print(f"   ‚ö†Ô∏è  T√¨m th·∫•y {negative_qty} gi√° tr·ªã Quantity √¢m")
    print(f"   V√≠ d·ª•: {df[df['Quantity'] < 0]['Quantity'].head().values}")
else:
    print("   ‚úì Kh√¥ng c√≥ gi√° tr·ªã Quantity √¢m trong d·ªØ li·ªáu")

# 3. Ki·ªÉm tra UnitPrice
print("\n3Ô∏è‚É£ UnitPrice:")
zero_price = (df['UnitPrice'] == 0).sum()
negative_price = (df['UnitPrice'] < 0).sum()

if zero_price > 0:
    print(f"   ‚ö†Ô∏è  T√¨m th·∫•y {zero_price} gi√° tr·ªã UnitPrice = 0")

if negative_price > 0:
    print(f"   ‚ö†Ô∏è  T√¨m th·∫•y {negative_price} gi√° tr·ªã UnitPrice √¢m")
    
if zero_price == 0 and negative_price == 0:
    print("   ‚úì UnitPrice h·ª£p l·ªá (kh√¥ng c√≥ 0 ho·∫∑c gi√° tr·ªã √¢m)")
```

## C√¢u 4: Range th·ªùi gian c·ªßa InvoiceDate

```python
# Chuy·ªÉn ƒë·ªïi InvoiceDate sang datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%d/%m/%Y', errors='coerce')

print("üìÖ Range th·ªùi gian c·ªßa InvoiceDate:\n")
print(f"Min: {df['InvoiceDate'].min().strftime('%d/%m/%Y')}")
print(f"Max: {df['InvoiceDate'].max().strftime('%d/%m/%Y')}")

# T√≠nh kho·∫£ng th·ªùi gian
date_range = (df['InvoiceDate'].max() - df['InvoiceDate'].min()).days
print(f"\n‚è±Ô∏è  Kho·∫£ng th·ªùi gian: {date_range} ng√†y ({date_range // 365} nƒÉm, {(date_range % 365) // 30} th√°ng)")

# Ki·ªÉm tra c√≥ ng√†y invalid kh√¥ng
invalid_dates = df['InvoiceDate'].isna().sum()
if invalid_dates > 0:
    print(f"\n‚ö†Ô∏è  T√¨m th·∫•y {invalid_dates} gi√° tr·ªã InvoiceDate kh√¥ng h·ª£p l·ªá")
```

## C√¢u 5: Ki·ªÉm tra d·∫•u hi·ªáu l·∫° trong c·ªôt quan tr·ªçng

```python
print("üî¨ Ki·ªÉm tra d·∫•u hi·ªáu l·∫° trong c·ªôt quan tr·ªçng:\n")
print("="*60)

# 1. Quantity √¢m
print("\n1Ô∏è‚É£ Quantity c√≥ √¢m kh√¥ng?")
negative_qty = (df['Quantity'] < 0).sum()
print(f"   T√¨m th·∫•y: {negative_qty} gi√° tr·ªã √¢m ({(negative_qty/len(df)*100):.2f}%)")
if negative_qty > 0:
    print(f"   Min Quantity: {df['Quantity'].min()}")
    print(f"   V√≠ d·ª• negative values: {df[df['Quantity'] < 0]['Quantity'].head(3).values}")

# 2. UnitPrice = 0
print("\n2Ô∏è‚É£ UnitPrice = 0 c√≥ t·ªìn t·∫°i kh√¥ng?")
zero_price = (df['UnitPrice'] == 0).sum()
print(f"   T√¨m th·∫•y: {zero_price} gi√° tr·ªã = 0 ({(zero_price/len(df)*100):.2f}%)")

# 3. UnitPrice √¢m
print("\n3Ô∏è‚É£ UnitPrice c√≥ √¢m kh√¥ng?")
negative_price = (df['UnitPrice'] < 0).sum()
print(f"   T√¨m th·∫•y: {negative_price} gi√° tr·ªã √¢m ({(negative_price/len(df)*100):.2f}%)")
if negative_price > 0:
    print(f"   Min UnitPrice: {df['UnitPrice'].min()}")

# 4. CustomerID tr·ªëng
print("\n4Ô∏è‚É£ CustomerID c√≥ nhi·ªÅu gi√° tr·ªã tr·ªëng kh√¥ng?")
missing_customer = df['CustomerID'].isna().sum()
print(f"   T√¨m th·∫•y: {missing_customer} gi√° tr·ªã NaN ({(missing_customer/len(df)*100):.2f}%)")

# 5. InvoiceNo ch·ª©a 'C' (cancel)
print("\n5Ô∏è‚É£ InvoiceNo c√≥ ch·ª©a ch·ªØ 'C' (cancel) kh√¥ng?")
cancel_invoices = df['InvoiceNo'].astype(str).str.contains('C', na=False).sum()
print(f"   T√¨m th·∫•y: {cancel_invoices} InvoiceNo ch·ª©a 'C' ({(cancel_invoices/len(df)*100):.2f}%)")

print("\n" + "="*60)
print("\nüìä T√ìM T·∫ÆT C√ÅC B·∫§T TH∆Ø·ªúNG:\n")

issues = []
if negative_qty > 0:
    issues.append(f"‚úì Quantity √¢m: {negative_qty} d√≤ng")
if zero_price > 0:
    issues.append(f"‚úì UnitPrice = 0: {zero_price} d√≤ng")
if negative_price > 0:
    issues.append(f"‚úì UnitPrice √¢m: {negative_price} d√≤ng")
if missing_customer > 0:
    issues.append(f"‚úì CustomerID tr·ªëng: {missing_customer} d√≤ng ({(missing_customer/len(df)*100):.2f}%)")
if cancel_invoices > 0:
    issues.append(f"‚úì InvoiceNo ch·ª©a 'C': {cancel_invoices} d√≤ng ({(cancel_invoices/len(df)*100):.2f}%)")

if not issues:
    print("‚úÖ Kh√¥ng t√¨m th·∫•y d·∫•u hi·ªáu l·∫° n√†o!")
else:
    for issue in issues:
        print(issue)
```
